```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#matplotlib plot bar plot  Requirement 4

ypos=np.arange(len(tuition_cost_by_dept_median["Department"]))
plt.barh(ypos,tuition_cost_by_dept_median["Cost"],color='rgbkymc')
plt.yticks(ypos,tuition_cost_by_dept_median["Department"])
plt.xlabel("Cost")
plt.title("Cost for each Department")
plt.show()
```

```python
#matplotlib plot bar plot Major/Degrees

ypos=np.arange(len(tuition_by_maj_degr["Major"]))
plt.barh(ypos,tuition_by_maj_degr["Degree"],color='rgbkymc')
plt.yticks(ypos,tuition_by_maj_degr["Major"])
#plt.legend(tuition_by_maj_degr["Degree"])
plt.xlabel("Degree")
plt.ylabel("Majors")
plt.title("Majors and Degrees")
plt.show()
```

```python
#matplotlib plot bar plot Department/Major

ypos=np.arange(len(tuition_by_dept_maj["Department"]))
plt.barh(ypos,tuition_by_dept_maj["Major"],color='rgbkymc')
plt.yticks(ypos,tuition_by_dept_maj["Department"])
plt.xlabel("Major")
plt.ylabel("Department")
plt.title("Majors and Department")
plt.show()
```
